Hello there, I wanted to show you a quick trick.
Too many times forgotten.
It will allows you to faster your preprocessing and model training.
For this, setting up the input type is important.
*pd.read_csv* does not do it by default :-)

### Usual way

```python
df_train = pd.concat([
    pd.read_csv("/kaggle/input/playground-series-s5e2/train.csv"),
    pd.read_csv("/kaggle/input/playground-series-s5e2/training_extra.csv"),
])

df_test = pd.read_csv("/kaggle/input/playground-series-s5e2/test.csv")
```

```python
df_total = pd.concat([df_train, df_test])
df_total.info()
```

```python
# Storing the memory usage.
start_memory_usage = 384.0
```

We can now improve it by setting the type of the columns.
Especially float ones can take many space.

After some EDA step, we also can notice that **Compartments** is actually a categorical column.

```python
# int16
print(f"Int16: {np.iinfo(np.int16)}")
print(f"Int32: {np.iinfo(np.int32)}")
print(f"Int64: {np.iinfo(np.int64)}")
print(f"Float16: {np.finfo(np.float16)}")
print(f"Float32: {np.finfo(np.float32)}")
print(f"Float64: {np.finfo(np.float64)}")
```

So watching the min and max values and doing some EDA, we can find the right type.

Changing 'object' to 'category' is a good start as well.

```python
categorical_column_names = df_total.columns[1:-2]
df_total["id"] = df_total["id"].astype("int32") # Even the max(id)=4_194_318 should fit.
df_total["Weight Capacity (kg)"] = df_total["Weight Capacity (kg)"].astype("float32") # Fits within float32  
df_total["Price"] = df_total["Price"].astype("float16") # Fits within float32
for column_name in categorical_column_names:
    df_total[column_name] = df_total[column_name].astype("category")
df_total.info()
```

```python
# Storing the memory usage.
end_memory_usage = 104.0

# Improvements.
improvment = (end_memory_usage - start_memory_usage) / start_memory_usage
print(f"Improvement: {100 * improvment:.2f}%.")
```

Improvement of 73%, not bad.

In this period where we need more and more computational power and enery to run the CPU, better take care about the consuptiom:)
